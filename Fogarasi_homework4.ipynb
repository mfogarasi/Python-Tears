```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
```

```python
#Throw some imaginary Darts to estimate a value for pi
def Darts(N):
    #Throw "darts" aka generate a two lists of random numbers [0,1) only 1st quadrant
    #list 1 is x, list 2 is y, each index is a "dart"
    x = np.random.rand(N) 
    y = np.random.rand(N)
    #print(x)
    #print(y)
    darts = np.vstack((x,y)).T
    #darts

    hits_x = []
    hits_y = []
    misses_x = []
    misses_y = []

    #loop through the points x^2 + y^2 if less than or equal to 1 "hit" else "miss"
    for index in range(len(darts)):
        if x[index]**2 + y[index]**2 <= 1:
            hits_x.append(x[index])
            hits_y.append(y[index])
        
        else:
            misses_x.append(x[index])
            misses_y.append(y[index])
            
    print('Number of hits: ', len(hits_x))
    print('Number of misses: ', len(misses_x))
    
    #Calculate estimate for pi using Area ratios, pi = (4*hits/N)
    pi = (4*len(hits_x))/len(x)
    print('Pi Estimate: ', pi)
    
    #Calculate the error from built in pi function 
    error = ((pi - math.pi)/math.pi) * 100
    print('Error: ',error,'%')

    #graph the darts on the board, hit as blue, miss as red
    plt.figure()
    plt.axis('square')
    plt.xlim((-1,1))
    plt.ylim((-1,1))
    circle1 = plt.Circle((0, 0), 1.0, fill=False)
    fig, ax = plt.gcf(),plt.gca() 
    ax.add_artist(circle1)
    #plot hit in blue and misses in red
    plt.scatter(hits_x, hits_y,  color = 'blue')
    plt.scatter(misses_x, misses_y, color = 'red')
```

## Estimating Pi
Area of 2x2 square = 4, r = 1, A=pi*r^2

Many hits (h) should eventually fill circle and be equal to area of 1st quadrant (because we are only generating numbers between 0 and 1). 

Therefore large N means h = (pi*r^2)/4 and r = 1 so h = pi/4. Area of the square  should eventually be equal to N or for 1st quadrant N = r^2. 

Make a ratio: hits/darts = pi/4

Solve for pi = (4*hits)/N

N = length of x or y coordinates array.

hits = length of hits array formed in loop by testing x^2 + y^2 < 1 and adding true to an array.
